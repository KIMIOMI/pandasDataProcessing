```python
import pandas as pd
```

```python
data = pd.read_excel('data.xlsx')
data
```

```python
원본이전카드 = data['이전 카드']
원본현재카드 = data['주소']
원본작성자 = data['작성자']
```

```python
원본작성자.unique()
```

```python
이전카드_개수 =[]
이전카드_앞주소제거=[]
제거값 = 'https://www.notion.so/'
for i in 원본이전카드:
    if type(i) == str:
        i = i.replace(제거값,'')
        주소 = i.split()
        a = len(주소)
    else:
        a = ''
        주소 = ''
    이전카드_개수.append(a)
    이전카드_앞주소제거.append(주소)

```

```python
현재카드_앞주소제거=[]
제거값1 = 'https://ageofzen.notion.site/'
제거값2 = 'https://www.notion.so/ageofzen/'
for i in 원본현재카드:
    if type(i) == str:
        주소 = i.replace(제거값1,'').replace(제거값2, '')
    else:
        주소 = ''
    현재카드_앞주소제거.append(주소)
    
```

```python
index = list(range(0,len(현재카드_앞주소제거)))
a = pd.Series(현재카드_앞주소제거)
b = pd.Series(이전카드_앞주소제거)
c = pd.Series(원본작성자)


new_data = pd.DataFrame(a, index, ['현재카드'])
new_data['이전카드'] = b
new_data['원본작성자'] = c

new_data
```

```python
new_data['모든이전카드'] = ''
new_data['모든기여자'] = ''
모든이전카드_개수 = []
모든이전카드 = []
에러카드 = []
모든기여자 = []

num = 0
for i in new_data.index:
    이전카드들 = new_data['이전카드'][i]
    작성자 = new_data['원본작성자'][i]
    모든기여자 = []
    모든기여자.append(작성자)

    if type(이전카드들) == list:
        모든카드 = 이전카드들.copy()
    else:
        모든카드 = ''

    if len(이전카드들) == 0:
        pass
    else:
        for 이전카드 in 이전카드들:
            쿼리결과 = new_data.query(f"현재카드 == '{이전카드}'")
            쿼리카드 = 쿼리결과['모든이전카드'].tolist()
            쿼리작성자 = 쿼리결과['모든기여자'].tolist()
            if len(쿼리카드) != 0:
                쿼리카드 = 쿼리카드[0]
                if len(쿼리작성자) != 0:
                    쿼리작성자 = 쿼리작성자[0]
                    for 작성자리스트 in 쿼리작성자:
                        if 작성자리스트 != '':
                            모든기여자.append(작성자리스트)
                    for 쿼리리스트 in 쿼리카드:
                        if 쿼리리스트 != '':
                            모든카드.append(쿼리리스트)
                else:
                    print(f"{i}번째 카드의 {이전카드} 작성자 오류")
            else:
                print(f'{i}번째 카드의 {이전카드}에서 쿼리 에러 발생')
                에러카드.append(이전카드)

    모든카드중복제거 = set(모든카드)
    모든카드 = list(모든카드중복제거)
    모든기여자 = list(set(모든기여자))

    new_data['모든이전카드'][i] = 모든카드
    new_data['모든기여자'][i] = 모든기여자
new_data['모든이전카드 개수'] = new_data['모든이전카드'].str.len()
에러카드 = list(set(에러카드))
new_data['모든기여자 수'] = new_data['모든기여자'].str.len()
        
```
